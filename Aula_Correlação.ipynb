### Carregar Base de Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# permitir que gr√°ficos possam ser impressos dentro do caderno jupyter.
%matplotlib inline
```

```python
dados = pd.read_excel('Lifexpec.xlsx', sheet_name='Life_Expectancy_Data')
```

```python
dados.info()
```

```python
# Selecionando os dados
dados_sel = dados.iloc[:, 3:21]
```

```python
dados_sel.head()
```

```python
# Calcular a matriz de correla√ß√£o
cor_matriz = dados_sel.corr()
cor_matriz
```

```python
# Arredondando para duas casas decimais
cor_matriz_arredondada = cor_matriz.round(2)

# Visualizando a matriz de correla√ß√£o arredondada
cor_matriz_arredondada
```

```python
# Criar uma m√°scara para ocultar os valores acima da diagonal
mask = np.triu(np.ones_like(cor_matriz, dtype=bool), k=1)

# Plotar a matriz de correla√ß√£o com a m√°scara
plt.figure(figsize=(15,10))
sns.heatmap(cor_matriz_arredondada, annot=True, cmap='coolwarm', mask=mask, vmin=-1, vmax=1, square=True, linewidths=.5)

# Ajustar r√≥tulos e exibir o gr√°fico
#plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
dados.columns
```

```python
# Grafico de dispers√£o
sns.scatterplot(x="hepatitis_B",y="diphtheria",
                data=dados)
# plt.title("Gr√°fico de Dispers√£o √≥bitos infantis por expectativa de vida")
plt.ylabel('Cobertura vacinal difteria e coqueluche')
plt.xlabel('Cobertura vacinal hepatite B')
plt.show()
```

```python
import numpy as np
from scipy import stats
```

```python
# Calculando o coeficiente de correla√ß√£o de Pearson e o p-valor
corr, p_value = stats.pearsonr(dados['hepatitis_B'], dados['diphtheria'])

# Imprimindo os resultados
print("Coeficiente de correla√ß√£o:", corr)
print("p-valor:", p_value)

```

```python
# Filtrar registros removendo os nulos nas duas colunas
dados2 = dados.dropna(how = 'any' , subset=[ 'hepatitis_B','diphtheria' ])
# Selecionar os dados das vari√°veis
X = dados2.hepatitis_B.values
Y = dados2.diphtheria.values
# Correla√ß√£o de Spearman
# H0 n√£o existe correla√ß√£o entre as vari√°veis
r, p = stats.spearmanr(X, Y)
# Mostrar o resultado
print ( 'Teste de Correla√ß√£o de Spearman' )
print ( 'Spearman r:', r.round(3))
print ( 'p-value:', p.round(3))

```

```python
# Boxplot mais elaborado que utiliza cores diferentes para cada categoria, adiciona uma legenda,
# um t√≠tulo, uma grade de fundo e a op√ß√£o notch=True para adicionar intervalos de confian√ßa nos quartis.

sns.boxplot(x='status', y='life_expectancy', hue='status', data=dados, palette='pastel', notch=False)

# Personalize r√≥tulos e t√≠tulo
plt.ylabel('Expectativa de vida')
#plt.xlabel('Status')
#plt.title('Boxplot da expectativa de vida por status')

# Exiba o gr√°fico
plt.show()
```
