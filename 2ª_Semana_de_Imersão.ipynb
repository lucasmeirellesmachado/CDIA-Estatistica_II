```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
base_med = pd.read_csv('https://raw.githubusercontent.com/lucasmeirellesmachado/CDIA-Estatistica_II/refs/heads/main/custo_medico_processamento.csv',
                       decimal = ',')
```

```python
#Tirar valores faltantes

#base_med.dropna() #Retira todas as linhas com dados faltantes

#base_med.dropna(axis = 1) #Retira todas as colunas com dados faltantes
```

```python
#Substituir os valores faltantes da coluna 'filhos' para '0'

base_med['filhos'] = base_med['filhos'].fillna(0)
```

```python
#base[(base['regiao'] == 'sudeste') & (base['filhos'].isnull())] ['filhos'] = 0
```

```python
#Substituir os valores faltantes de dievrsas colunas

base_med = base_med.fillna ({'fumante': 'nao',
                         'filhos': 0,
                         'doencas_cronicas': 0})
```

## Label enconding ##

```python
from sklearn.preprocessing import LabelEncoder #variáveis qualitativas nominais
```

```python
le_fumante = LabelEncoder()
```

```python
base_med['fumante_cod'] = le_fumante.fit_transform(base_med['fumante'])
```

```python
base_med[['fumante', 'fumante_cod']].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder #variáveis qualitativas ordinais
```

```python
oe_regiao = OrdinalEncoder(categories = [['nordeste',
                                          'noroeste',
                                          'sudeste',
                                          'sudoeste']])
```

```python
base_med['regiao_cod'] = oe_regiao.fit_transform(base_med[['regiao']])
```

```python
base_med[['regiao', 'regiao_cod']].value_counts()
```

## Dummização de dados ##

```python
dados = pd.get_dummies(base_med, columns = ['fumante'],
                       dtype=int, #Atentar para a multicolinearidade dessa variável, pois se trata da mesma, embora estejam em duas colunas
                       drop_first=True) # Retira uam das colunas para evitar a multicolinearidade
```
